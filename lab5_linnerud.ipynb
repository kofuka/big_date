```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split

# Разделение данных на признаки (X) и целевую переменную (y)
X, y = linnerud.data, linnerud.target

# Разделение данных на обучающий и тестовый наборы
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Создание и обучение модели Ridge регрессии
ridge = Ridge(alpha=1.0)
ridge.fit(X_train, y_train)

# Вывод коэффициентов модели Ridge
print("Коэффициенты Ridge:")
print(ridge.coef_)

# Оценка производительности модели на обучающем и тестовом наборах
train_score = ridge.score(X_train, y_train)
test_score = ridge.score(X_test, y_test)
print("Оценка на обучающем наборе: {:.2f}".format(train_score))
print("Оценка на тестовом наборе: {:.2f}".format(test_score))
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split

# Разделение данных на признаки (X) и целевую переменную (y)
X, y = linnerud.data, linnerud.target

# Разделение данных на обучающий и тестовый наборы
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Создание и обучение модели Lasso регрессии
lasso = Lasso(alpha=1.0)
lasso.fit(X_train, y_train)

# Вывод коэффициентов модели Lasso
print("Коэффициенты модели Lasso:")
print(lasso.coef_)

# Оценка производительности модели на обучающем и тестовом наборах
train_score = lasso.score(X_train, y_train)
test_score = lasso.score(X_test, y_test)
print("Оценка на обучающем наборе: {:.2f}".format(train_score))
print("Оценка на тестовом наборе: {:.2f}".format(test_score))
```
